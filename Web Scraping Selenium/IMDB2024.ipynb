```python
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.common.action_chains import ActionChains
from selenium.webdriver.common.keys import Keys
import time
```

```python

driver = webdriver.Chrome()

driver.get("https://www.imdb.com/search/title/?title_type=feature&release_date=2024-01-01,2024-12-31&genres=adventure")

time.sleep(3)
genre = "Adventure"

def click_load_more():
    try:
        load_more_button = driver.find_element(By.XPATH, "//span[contains(@class,'ipc-see-more__text')]")

        ActionChains(driver).move_to_element(load_more_button).perform()
        load_more_button.click()
        time.sleep(3)
        return True
    except Exception as e:
        print("Error clicking 'Load More' bu-tton:", e)
        return False
while click_load_more():
    print("Clicked Load More")
print("No more content to load.")
```

```python
movies = driver.find_elements(By.XPATH, "//li[contains(@class,'ipc-metadata-list-summary-item')]")
movie_data = []
for movie in movies:
    try:
        title = movie.find_element(By.XPATH, ".//h3[contains(@class,'ipc-title__text')]").text
    except:
        title = "N/A"

    try:
        voting = movie.find_element(By.XPATH,".//span[contains(@class,'ipc-rating-star--voteCount')]" ).text  
    except:
        voting = "N/A"
    try:
        rating = movie.find_element(By.XPATH, './/span[contains(@class,"ipc-rating-star--rating")]').text
    except:
        rating = "N/A"

    try:
        metadata = movie.find_elements(By.XPATH, ".//span[contains(@class,'dli-title-metadata-item')]")
        duration = metadata[1].text if len(metadata) > 1 else "N/A"


#duration = movie.find_element(By.XPATH, ".//span[contains(@class,'dli-title-metadata-item')]").text
    except:
        duration = "N/A"

    movie_data.append({
        "Title": title,
        "Genre": genre,
        "Duration": duration,
        "Rating": rating,
        "voting": voting,
    })

# Print scraped data
for movie in movie_data:
    print(movie)

# Close driver
#driver.quit()
```

```python
import pandas as pd

# Convert list to DataFrame
df = pd.DataFrame(movie_data)

# 1. Remove leading numbers from Title (e.g., "1. IF" → "IF")
df["Title"] = df["Title"].str.replace(r"^\d+\.\s*", "", regex=True)

# 2. Remove parentheses from Voting column
df["voting"] = df["voting"].str.replace(r"[()]", "", regex=True)

# 3. Reset index to start from 1
df.index = df.index + 1

# Save to CSV
df.to_csv(r"D:\GUVI\IMDB2024\animation.csv")

print("CSV file saved successfully in D:\\GUVI")
```
